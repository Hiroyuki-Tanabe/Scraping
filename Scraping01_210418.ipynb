```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import chromedriver_binary
import time
import requests, json, datetime
from bs4 import BeautifulSoup
import urllib.request as req
import csv
import numpy as np
import pandas as pd
```

```python
# 空のnumpy配列（300行,2列）を用意
array = [[0 for i in range(2)] for j in range(300)]
# for文に突っ込む
for num in range(300):
    print(num)
    # 通常処理
    try:
        url= "https://www.aizukanko.com/spot/{}"
        html = requests.get(url.format(num))
        soup = BeautifulSoup(html.content, 'html.parser')
        # タイトル・住所を取得
        title = soup.find_all("h1", class_="title")
        address = soup.find_all("td") [0]
        # テキストのみ抽出し、配列に格納
        for t in title:
            array[num][0] = t.getText()
        array[num][1] = address
        print( t.getText(), address)
    # 例外処理
    except:
        print("データなし")
        array[num][0] = "データなし"
        array[num][1] = "データなし"
    time.sleep(3)
```

```python
array
```

```python
# csvに書き出し
with open('aizu_spot.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerows(array)
```
